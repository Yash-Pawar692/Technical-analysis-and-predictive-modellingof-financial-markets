```python
#TCS
df_TCS = TCS.history(period='5y')[['Open', 'High', 'Low', 'Close', 'Volume']]
```

```python
plt.style.use('seaborn-pastel')
plt.style.use('seaborn')
plt.style.available
df_TCS
```

```python
df_TCS['SMA_30'] = df_TCS['Close'].rolling(window=30).mean()
df_TCS['SMA_50'] = df_TCS['Close'].rolling(window=50).mean()
df_TCS['SMA_100'] = df_TCS['Close'].rolling(window=100).mean()
df_TCS['SMA_200'] = df_TCS['Close'].rolling(window=200).mean()
df_TCS['SMA_vol'] = df_TCS['Volume'].rolling(window=50).mean()
fig = make_subplots(specs=[[{"secondary_y": True}]])
fig.add_trace(
    go.Scatter(x=df_TCS.index, y=df_TCS['Close']),
    secondary_y=False,
)

fig.show()
```

```python
#STMA
df_TCS_365 = df_TCS[-365:]
fig1 = make_subplots(specs=[[{"secondary_y": True}]])
fig1.add_trace(
    go.Scatter(x=df_TCS_365.index, y=df_TCS_365['Close'], name='Close'),
    secondary_y=False,
)
fig1.add_trace(
    go.Scatter(x=df_TCS_365.index, y=df_TCS_365['SMA_30'], name='30 day SMA'),
    secondary_y=False
)
```
